# Reverbetation Time Data Analysis using Schroders Integral

### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import librosa as lr
import librosa.display as ld
import scipy.signal as sc
import matplotlib.colors as mcolors
from scipy.optimize import curve_fit
```

### Main Script

```python
x,sr = lr.load('bedroom_RIR.wav')
t = np.arange(len(x))/sr


#Analytical signal Removes an phase information
x_anyl = abs(sc.hilbert(x))/np.max(abs(x))

#Take the log to get the power singal
x_log = 20.0*np.log10(x_anyl)

window_size = 50
ma_filter = np.ones(window_size) / window_size
x_ma = np.convolve(x_anyl, ma_filter, mode='same') #Filter the signal to remove noise


N = 7000  # desired number of samples for integration
x_ma1 = x_ma[:N]


#Integrate using Schroders integral
sch = 10.0 * np.log10(np.cumsum(x_ma1[::-1]**2)[::-1] / np.max(np.cumsum(x_ma1[::-1]**2)))
```

### Plotting

```python
plt.figure(figsize = (20,8))
plt.plot(t,x,'b')
plt.plot(t,x_anyl,'r')
plt.xlim(0,0.04)


plt.figure(figsize = (20,8))
#plt.plot(t,x_log,'b',label = 'Audio Signal')
plt.plot(t,20*np.log10(x_ma),'r',linewidth=3.5,label="Windowed Signal")
plt.plot(t,20*np.log10(x),'b',label='Audio Signal')
plt.plot(t[:N],sch,'xkcd:spearmint',linewidth=3.5,label = "Schroders Integral Curve")
plt.axhline(y=-60, color='k',ls='-.')
plt.text(0.1, -62, '-60 dB', ha='right', va='top', color='k')
plt.axhline(y=-30, color='k',ls='-.')
plt.text(0.5, -32, '-30 dB', ha='right', va='top', color='k')
plt.axhline(y=-20, color='k',ls='-.')
plt.text(0.5, -22, '-20 dB', ha='right', va='top', color='k')
plt.axhline(y=-10, color='k',ls='-.')
plt.text(0.5, -12, '-10 dB', ha='right', va='top', color='k')
plt.grid(True,which="both",ls="-")

plt.title("Energy vs Time Graph")
plt.xlabel("Time [s]")
plt.ylabel("Energy [dB]")
plt.legend()

## Run curve fit on the integral
def func(x,m,b):
    return m*x+b

popt,pconv = curve_fit(func,t[:N],sch)
m,b = popt
y = m * t + b
x_intersection = (-60-b)/m

plt.plot(t[:2*N],y[:2*N],color="tab:orange",ls="dotted",linewidth = 3.5,label = "Curve Fit")
plt.legend()

print(f"RT60 = {-60/m:.7f} sec")
print(f"RT30 = {-30/m:.7f} sec")
print(f"RT20 = {-20/m:.7f} sec")
print(f"RT10 = {-10/m:.7f} sec")
```
